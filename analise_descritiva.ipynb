```python
import pandas as pd
import sqlite3

# Função para carregar os dados da tabela "expectativa_mercado"
def carregar_dados_banco():
    con = sqlite3.connect("src/datasets/etlbcb.db")
    query = "SELECT * FROM expectativa_mercado"
    df = pd.read_sql(query, con)
    con.close()
    return df

# Função para gerar o relatório de análise descritiva
def gerar_analise_descritiva_separada(df: pd.DataFrame):
    """
    Função para gerar análise descritiva das colunas numéricas do DataFrame.
    A análise incluirá média, mediana, desvio padrão, mínimo, máximo e quantis,
    gerando um relatório separado para cada uma das estatísticas.
    """
    # Selecionar apenas as colunas numéricas para análise
    df_numerico = df.select_dtypes(include='number')
    
    # Gerar e exibir o relatório para cada estatística separadamente
    
    # Média
    media = df_numerico.mean()
    print("\nMédia das Colunas Numéricas:")
    print(media)
    
    # Mediana
    mediana = df_numerico.median()
    print("\nMediana das Colunas Numéricas:")
    print(mediana)
    
    # Desvio Padrão
    desvio_padrao = df_numerico.std()
    print("\nDesvio Padrão das Colunas Numéricas:")
    print(desvio_padrao)
    
    # Mínimo
    minimo = df_numerico.min()
    print("\nMínimo das Colunas Numéricas:")
    print(minimo)
    
    # Máximo
    maximo = df_numerico.max()
    print("\nMáximo das Colunas Numéricas:")
    print(maximo)
    
    # Quantis (25%, 50%, 75%)
    quantis = df_numerico.quantile([0.25, 0.5, 0.75])
    print("\nQuantis (25%, 50%, 75%) das Colunas Numéricas:")
    print(quantis)
    
    # Contagem de valores nulos
    valores_nulos = df_numerico.isnull().sum()
    print("\nValores Nulos nas Colunas Numéricas:")
    print(valores_nulos)

# Carregar os dados da tabela "expectativa_mercado"
df = carregar_dados_banco()

# Gerar e exibir a análise descritiva separada
gerar_analise_descritiva_separada(df)
```

```python
# Função para calcular a média
def calcular_media(df: pd.DataFrame):
    df_numerico = df.select_dtypes(include='number')  # Selecionando apenas as colunas numéricas
    media = df_numerico.mean()
    return media


media = calcular_media(df)
print("\nMédia das Colunas Numéricas:")
print(media)
```

```python
# Função para calcular a mediana
def calcular_mediana(df: pd.DataFrame):
    df_numerico = df.select_dtypes(include='number')
    mediana = df_numerico.median()
    return mediana


mediana = calcular_mediana(df)
print("\nMediana das Colunas Numéricas:")
print(mediana)
```

```python
# Função para calcular o desvio padrão
def calcular_desvio_padrao(df: pd.DataFrame):
    df_numerico = df.select_dtypes(include='number')
    desvio_padrao = df_numerico.std()
    return desvio_padrao


desvio_padrao = calcular_desvio_padrao(df)
print("\nDesvio Padrão das Colunas Numéricas:")
print(desvio_padrao)
```

```python
# Função para calcular o mínimo e máximo
def calcular_min_max(df: pd.DataFrame):
    df_numerico = df.select_dtypes(include='number')
    minimo = df_numerico.min()
    maximo = df_numerico.max()
    return minimo, maximo


minimo, maximo = calcular_min_max(df)
print("\nMínimo das Colunas Numéricas:")
print(minimo)
print("\nMáximo das Colunas Numéricas:")
print(maximo)
```

```python
# Função para calcular os quantis (25%, 50%, 75%)
def calcular_quantis(df: pd.DataFrame):
    df_numerico = df.select_dtypes(include='number')
    quantis = df_numerico.quantile([0.25, 0.5, 0.75])
    return quantis


quantis = calcular_quantis(df)
print("\nQuantis (25%, 50%, 75%) das Colunas Numéricas:")
print(quantis)
```

```python
# Função para contar valores nulos
def contar_valores_nulos(df: pd.DataFrame):
    df_numerico = df.select_dtypes(include='number')
    valores_nulos = df_numerico.isnull().sum()
    return valores_nulos


valores_nulos = contar_valores_nulos(df)
print("\nValores Nulos nas Colunas Numéricas:")
print(valores_nulos)
```

```python
import matplotlib.pyplot as plt

# Função para gerar visualização (histograma) de uma coluna específica
def visualizacao(df: pd.DataFrame, coluna: str):
    df_numerico = df.select_dtypes(include='number')
    df_numerico[coluna].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
    plt.title(f"Distribuição da {coluna}")
    plt.xlabel(coluna)
    plt.ylabel("Frequência")
    plt.show()


visualizacao(df, 'Media')
```

```python
# Função para contar valores categóricos
def contar_valores_categoricos(df: pd.DataFrame, coluna: str):
    categorias = df[coluna].value_counts()
    return categorias


categorias = contar_valores_categoricos(df, 'Indicador')
print("\nDistribuição de Indicador:")
print(categorias)
```
